#### Step 7: Query Json fields

```python
## As you can tell from the previous step, a few fields store json type of data.

## genre: [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
## production_companies: [{"name": "Ingenious Film Partners", "id": 289}, {"name": "Twentieth Century Fox Film Corporation", "id": 306}, 
## {"name": "Dune Entertainment", "id": 444}, {"name": "Lightstorm Entertainment", "id": 574}]

## Hive has a function get_json_object that can parse json data. Unfortunately, it does not accept json data that is an array at the root (which is our case). 
## To circumvent that, we can transform the json data to one that has a root field with the array as its value. e.g. {"root":[{"id": 28, "name": "Action"}, 
## {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]}. 

## This transformation can be done by concat('{"root":',field,'}').

## 16. With the above tip, query the movies table to show title and the first production_company, limiting results to 10. Report Result

0: jdbc:hive2://> SELECT title, GET_JSON_OBJECT(CONCAT('{"root":', production_companies, '}'), '$.root.name[0]') AS production_company FROM movies LIMIT 10;
OK
+-------------------------------------------+--------------------------+--+
|                   title                   |    production_company    |
+-------------------------------------------+--------------------------+--+
| Avatar                                    | Ingenious Film Partners  |
| Pirates of the Caribbean: At World's End  | Walt Disney Pictures     |
| Spectre                                   | Columbia Pictures        |
| The Dark Knight Rises                     | Legendary Pictures       |
| John Carter                               | Walt Disney Pictures     |
| Spider-Man 3                              | Columbia Pictures        |
| Tangled                                   | Walt Disney Pictures     |
| Avengers: Age of Ultron                   | Marvel Studios           |
| Harry Potter and the Half-Blood Prince    | Warner Bros.             |
| Batman v Superman: Dawn of Justice        | DC Comics                |
+-------------------------------------------+--------------------------+--+
10 rows selected (0.373 seconds)
```
