```python
def total_consumer_credit():
    
    # read in the URL and grab the Beautiful soup object
    r = urllib.request.urlopen('https://www.federalreserve.gov/releases/g19/current/default.htm')
    soup = BeautifulSoup(r, "lxml")
    
    # collect all the information needed to create the dataframe, indexes, columns, and data rows
    columns = soup.table.text.split()[17:30]
    data_row1 = soup.table.text.split()[139:152]
    data_row2 = soup.table.text.split()[154:167]
    index1 = soup.table.text.split()[138]
    index2 = soup.table.text.split()[152] + " " + soup.table.text.split()[153]
    
    # create the initial dataframe with the data rows, transponse them, and then set the column names
    soup_df = DataFrame(data=list(zip(data_row1,data_row2))).T
    soup_df.columns = columns
    
    # change the index names to match the output in the exercise example
    first_index = soup_df.index[0]
    second_index = soup_df.index[1]

    soup_df = soup_df.rename(index={first_index: index1})
    soup_df = soup_df.rename(index={second_index: index2})
    
    return soup_df
```
