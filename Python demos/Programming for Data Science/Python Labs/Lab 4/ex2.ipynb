```python
#def golden_ratio

def fibonacci_recursive(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)
```

```python
def golden(num):
    g = [1,1]
    for i in range(num):
        g.reverse() # calculate 1/g
        g[0] += g[1] # add 1
    return g[0] / g[1]
```
