```python
import pandas as pd
from pandas import Series, DataFrame
import datetime as dt
```

```python
def char_diff(dictionary):
    
        word1 = dictionary.get('misspelled word')
        word2 = dictionary.get('corrected word')
        
        word1_length = len(word1)
        word2_length = len(word2)
        
        if word1_length < word2_length:
            test_range = word1_length  
        elif word2_length < word1_length:
            test_range = word2_length
        elif word1_length == word2_length:
            test_range = word1_length
    
        count = 0
        for i in range(test_range):
            if word1[i] != word2[i]:
                count+=1
        return count

def avg_char_diff(filename, n):
    
    misspellings_df = pd.read_csv(filename)
    
    misspellings_df['diff'] = misspellings_df.apply(char_diff, axis = 1)
    
    misspellings_df = misspellings_df.groupby('corrected word').mean()
    
    misspellings_df = misspellings_df.loc[misspellings_df['diff'] >= n]
    
    return pd.Series(misspellings_df['diff'])
```
