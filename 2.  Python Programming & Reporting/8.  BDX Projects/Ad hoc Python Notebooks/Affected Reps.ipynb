```python
teams = plus.read_google_sheet('114yDQZ7Fi6AhXBg2IhO3NZxEmAcvL12AuyXSIsZN38U',intype='key',header=True,index=False,sheet='Sheet1')
```

```python
applywteam = applybylocale.merge(teams,how='left',left_on='service rep',right_on='username')
```

```python
def team(x):
    if x['service rep']=='CAsharedsupport':
        return 'CA Shared Support'
    if x['service rep']=='APACSharedSupport':
        return 'APAC Shared Support'
    if x['service rep']=='DCASharedSupport':
        return 'DCA Shared Support'
    if x['service rep']=='EMEAsharedsupport':
        return 'EMEA SharedSupport'
    if x['service rep']=='agencyhostedsupport':
        return 'Agency Hosted Job Support'
    if x['service rep']=='sharedclientsupport':
        return 'Shared Client Support'
    if x['service rep']=='smbsupp':
        return 'Small Business Support'
    else:
        return x['Team']
    
applywteam['Team']=applywteam.apply(team,axis=1)
```
