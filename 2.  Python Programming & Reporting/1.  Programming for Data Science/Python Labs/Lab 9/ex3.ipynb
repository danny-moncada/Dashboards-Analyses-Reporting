```python
import pandas as pd
from pandas import Series, DataFrame
import datetime as dt

def four_hour_temp(filename, date):
    
    weather_df = pd.read_csv(filename, parse_dates = [["Date", 'Time']], index_col = ['Date_Time'],
                     na_values = 'M')
    
    return weather_df.loc[date].resample('4H')['DryBulbCelsius'].mean()
```
