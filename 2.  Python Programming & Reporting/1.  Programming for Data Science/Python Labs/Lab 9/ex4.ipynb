```python
import pandas as pd
from pandas import Series, DataFrame
import datetime as dt

def adjusted_gdp(usgs, cpi, start, end):
    
    start = str(start)
    end = str(end)
    
    us_gdp = pd.read_csv(usgs, index_col = 0, parse_dates = True)
    us_gdp = us_gdp[start : end]
    us_gdp = us_gdp.shift(periods=1, freq='A', axis = 0)
    
    cpiai = pd.read_csv("cpiai.csv", index_col = 0, parse_dates = True)
    cpiai = cpiai[start : end]
    cpiai = cpiai.groupby(cpiai.index.year).mean()
    cpiai.index = pd.to_datetime(cpiai.index.astype(str))
    cpiai = cpiai.shift(periods=1, freq='A', axis = 0)
    
    adjusted_gdp = us_gdp.merge(cpiai, left_index=True, right_index=True)
    adjusted_gdp['adjusted_gdp'] = round(adjusted_gdp['GDP-US $ billion nominal'] / adjusted_gdp['Index'] * 100, 2)
    return adjusted_gdp['adjusted_gdp']
    
```
