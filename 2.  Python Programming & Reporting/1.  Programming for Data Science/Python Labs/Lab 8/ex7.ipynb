```python
import pandas as pd
from pandas import Series, DataFrame
import datetime as dt
import time

def most_delayed_flights(filename, percentage):
    
        flight_delays_df = pd.read_csv(filename)
        
        flight_delays_df['FL_DATE'] = pd.to_timedelta(flight_delays_df['FL_DATE'], unit = 'D') + dt.datetime(1899, 12, 30)
        flight_delays_df.set_index(['FL_DATE'], inplace = True)
        
        eligible_flights = flight_delays_df.groupby(['CARRIER', 'FL_NUM']).resample('W-Sun')['Flight Status'].count().reset_index()
        eligible_flights = eligible_flights[eligible_flights['Flight Status'] >= 3]
        
        flights = eligible_flights['FL_NUM']
        
        flight_delays_df = flight_delays_df[flight_delays_df['FL_NUM'].isin(flights)]
        
        flight_delays_cnt = pd.crosstab([flight_delays_df.CARRIER, flight_delays_df.FL_NUM] , flight_delays_df['Flight Status']).reset_index()

        flight_delays_cnt['sum of flights'] = flight_delays_cnt['ontime'] + flight_delays_cnt['delayed']

        flight_delays_cnt = flight_delays_cnt[flight_delays_cnt['sum of flights'] >= 3]

        flight_delays_cnt['mean'] = flight_delays_cnt['delayed'] / flight_delays_cnt['sum of flights']

        flight_delays_cnt = flight_delays_cnt[flight_delays_cnt['mean'] > percentage]

        return flight_delays_cnt.groupby(['CARRIER', 'FL_NUM'])['mean'].sum()
```

```python
def loan_by_zip(filename, n, loan_approval_rate):
    
    bank_df = pd.read_csv(filename)

    zip_code_count = DataFrame(bank_df.groupby('ZIP Code')['Personal Loan'].count())
    zip_code_count.rename(index=str, columns = {'Personal Loan':'size'}, inplace = True)

    approved_loans = DataFrame(bank_df.groupby('ZIP Code')['Personal Loan'].sum())
    approved_loans.rename(index=str, columns = {'Personal Loan':'approved loans'}, inplace = True)

    merged_df = zip_code_count.merge(approved_loans, left_index = True, right_index = True)

    merged_df['mean'] = round(merged_df['approved loans'] / merged_df['size'], 1).values

    return merged_df[(merged_df['size'] >= n) & (merged_df['mean'] >= loan_approval_rate)][['mean', 'size']]
```
