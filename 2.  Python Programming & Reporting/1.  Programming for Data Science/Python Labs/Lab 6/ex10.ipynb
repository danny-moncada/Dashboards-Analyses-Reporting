```python
import pandas as pd

from pandas import Series, DataFrame

def new_to_list(filename, year):
    
    fortune_500 = pd.read_csv(filename)
    
    year_before = fortune_500[fortune_500["Year"] == year - 1]["Company"]
    
    year_to_search = fortune_500[fortune_500["Year"] == year]["Company"]
    
    return year_to_search[~year_to_search.isin(year_before)].sort_values().values
```

```python
new_to_list("fortune500.csv", 1995)
```

```python
new_to_list("fortune500.csv", 2006)
```
