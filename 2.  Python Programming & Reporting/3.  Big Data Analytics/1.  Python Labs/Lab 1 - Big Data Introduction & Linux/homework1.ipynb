Big Data Use Case: Find one real-world use case, use the framework we introduce in class (the pyramid and 5 V’s big data) to discuss what characteristics of data in this use case make it a big data application, what are the technical capabilities being used, and what values it provides to the organization.

<p>  Domino's Pizza went through an extensive exercise in the early 2010's to start using technology to their advantage and improve their business processes.  With their online pizza-ordering Web site (launched in 2005), and with an iPhone-optimized mobile ordering system (launched in 2009), they decided to use customer information and insights that they were gleaming every day to make their business practices more efficient. </p>

Here are the characteristics that make this a big data use case:

- <b>Volume</b>: Dominos had over 9000 franchise stores by the end of 2010, in locations spanning North America, Central America, South America, Europe, East Asia, and Australia. [Cherian & Merchant, pg. 2]  This is an absolute treasure trove of information from a variety of places around the world.
- <b>Velocity</b>: Dominos processes several hundred thousand transactions per day. [Cherian & Merchant, pg. 7]  They also get real-time updates from customers and from the pizza tracking app.


The technical capabilities (<b>Variety</b>) being used were as follows:
- <span style="color:red">Leaderboard</span>: A ranking tool in every store that allowed managers to access performance analytics and operational metrics.  This same data was available at Domino's headquarters and via mobile phones.  This ran on live data used and updated by PULSE and Pizza Tracker. [Cherian & Merchant, pg. 5]
- <span style="color:red">Pizza Tracker</span>: A Web application developed for customer engagement, built on top of the PULSE system created in 2003.  It allowed customers to view the status of their order at a very high level of data granularity, to the point of being able to know the name of the person who assembled their pizza. [Cherian & Merchant, pg. 5]
- <span style="color:red">PULSE</span>: A point-of-sale system created in 2003 that contained day-to-day operational data.
- <span style="color:red">Web 2.0 platform</span>: An internal employee platform for wikis, blogs, and discussion forums.  This encouraged their employees to resource share and provided a platform to share best practices among their community. [Cherian & Merchant, pg. 6]
- <span style="color:red">"Think Oven"</span>: A Facebook application to capture ideas from Facebook fans.  They were able to grow their fan base to over 1 million within 90 days. [Cherian & Merchant, pg. 6]

The <b>Value</b> that it provided to the organization was three-fold.
- a. <b>Competition as a Motivator</b>: The Leaderboard exposed stores' performance in terms of their neighboring stores and across the company and created a social incentive to perform.  The social pressure was a more effective incentive for stores to do well, since a manager could get on a phone and gloat over a nearby store manager for performing better.  This made it easier to motivate employees to provide faster "out the door" times, i.e. getting pizza out of the store quickly and into a customer's home.
- b. <b>Instant Customer Feedback</b>: Historically, it was rare for customers to contact the company directly to provide a complaint with bad service; instead, they would simply go with a competitor the next time they wanted pizza.  With the Pizza Tracker, it was much easier for customers to rate Domino's products and their overall experience.  To address the high volume, feedback was routed to store managers who could contact dissastified customers to resolve their issues right away.  Domino's was able (eventually) to improve their overall customer satisifcation metrics - this is a *huge* for a company who are operating in a very challenging landscape like food delivery.
- c. <b>A More Personal Customer Experience</b>: They made the customer experience, which occurred millions of times per week, more personalized.  They started by including the name of the employee making the pizza as well as the delivery driver.  Employees were able to benefit from positive recognition from customers, because customers felt more comfortable knowing the name of the driver going to their home and would reward drivers with a better tip as a result.  In addition, the Pizza Tracker allowed their customers to check the status of their order from the time it was placed until it arrived and open a line of communication to learn about their customer's concerns and suggestions.  They improved their speed of delivery, operational transparency, and sensitivity to customer expectations and emotions. [Cherian & Merchant, pg. 6]

<p>
I think this point is important because Variety is a huge component of Big Data; Variety provides different perspectives on a particular subject or measurement and allows business partners/managers to make better decisions.  For example, if you are a sales manager at Target and you are gearing up for a new product launch, you wouldn't just want to look at sales figures for similar products that you already have in your inventory; you would also look at sales figures for similar products for your <b>competitors</b>.  You can build a much better projection of how well the product is going to do if you aren't basing your sales projections on just internal data, but also including data from external sources in your model(s).
</p>

<p>b. "A key skill involves explaining big data outcomes to executives--in visual displays or verbal narrative.  Even among data scientists, several interviewees commented that their quantatitve people need to 'tell a story with data,' and relate well to decision-makers." [Davenport & Dyche, pg. 14]
</p>
<p>

One of the biggest challenges for any analyst is taking a large data set and extracting meaningful information from it.  For example, we perform a lot of operational reporting around the headcount and full-time equilavency of each department, college, and at the campus level for the entire University of Minnesota.  This data can be very complex because of the multiple changes that happen to our university workforce <i>every</i> day, and making a visualizations that are easy to view and consume can be extremely challenging, especially for many of our senior leaders that are used to simple Excel files and pivot tables.
