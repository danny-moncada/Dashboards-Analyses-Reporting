```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sun_country_df = pd.read_csv("SunCountry.csv", low_memory = False)
```

```python
sun_country_df.tail(20)
```

```python
sun_country_df['key'] = sun_country_df['TicketNum'].astype(str) + sun_country_df['CouponSeqNbr'].astype(str)
```

```python
sun_country_df[sun_country_df['TicketNum'] == 3377336401688].sort_values('CouponSeqNbr')
```

```python
ticket_oneway = sun_country_df.groupby('TicketNum')['CouponSeqNbr'].max().reset_index()
```

```python
ticket_oneway.groupby('CouponSeqNbr').count().reset_index().to_csv('ticketoneway.csv')
```

```python
sun_country_df.groupby('TicketNum')['CouponSeqNbr'].max()
```

```python
end_cities = sun_country_df[['TicketNum', 'ServiceEndCity']]
```

```python
one_way_city_dest = ticket_oneway.merge(end_cities, how = 'left', left_on = 'TicketNum', right_on = 'TicketNum' )
```

```python
one_way_city_dest[one_way_city_dest['CouponSeqNbr'] == 1].to_csv('one_way_end_city.csv')
```

```python
count_df = sun_country_df.groupby('EncryptedName')['BookingChannel'].count()
```

```python
count_df = count_df.reset_index()
```

```python
histo = count_df.groupby('BookingChannel').count()
```

```python
plt.hist(histo)
plt.show()
```

```python
histo.to_csv('histoftrip.csv')
```
