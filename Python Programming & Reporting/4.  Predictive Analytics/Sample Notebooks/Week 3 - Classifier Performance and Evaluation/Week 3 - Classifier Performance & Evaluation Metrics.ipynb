### Bad Positives & Harmless Negatives<br>

* Positive examples are worthy of attention or *alarm*, negative example is benign or uninteresting<Br>
* i.e. medical test tries to detect disease or unusual condition;<br>
** If test is +, then disease or condition is present<br>
** If test is -, then no cause for concern or alarm<br>
<br>
<br>
* i.e. Fraud detector finds unusual activity on customer account, this is +<br>
* If fraud detector finds nothing, then -<br>
<br>

### A classifier sifts through a large population consisting mostly of negatives -  the un-interesting cases - looking for a small numer of positive instances.<br>

* *Number* of mistakes made on negative examples (*false positive* errors) may dominate but the *cost* of each mistake made on positive example (*false negative* error) will be higher.

accuracy = # of correct decisions made / total # of decisions made (or 1 - *error rate*)

### Confusion Matrix<br>

*n* classes is *n* x *n* matrix with **columns** as actual classes and **rows** labeled with predicted class

### Problems with Unbalanced Classes<br>

* Classifiers are often used to sift through large population of normal or uninteresting entities to find relatively small number of unusual ones<br>
** Defrauded customers, defective parts, target customers responding to offers<br>
* Class distribution is *skewed*<br>

* Side effect is as class distribution becomes more skewed, evaluation based on accuracy gets worse.<br>

* I.e. domain where classes are in 999:1 ratio.  Simple majority rule would give 99.9% accuracy, which is not satisfactory.

#### Let's take case above: A falsely predicts customers will churn when they will not, while B makes many opposite errors of predicting that customers will not churn when they will.  When applied to original, unbalanced population of customers, A's accuracy plummets to 37% while B rises to 93%.

#### So A, because it is classifying differently, will be way off when actually applied to the whole pop.  Requires better classification.

### Key Analytical Framework: Expected Value<br>

Expected value:<br>
i. Structure of problem<br>
ii. Elements of analysis that can be extracted from data<br>
iii. Elements of analysis that need to come from outside sources (business knowledge/SME)<br>

<br>
* It is the weighted average of values of different possible outcomes, with the weight given to each value based on probability of it actually ocurring.<br>
<br>

**Expected profit calculation**: weights heavily highly likely levels of profit, while unlikely levels of profit have little weight.
