# Measures of Error<br>




* MAE (Mean Absolute Error)


* MAPE (Mean Absolute Percentage Error)


* RMSE (Root Mean Squared Error)


* Total SEE (Sum of Squared Errors)


* Average Error

<br>** Systematic over/under prediction


* Most popular: RMSE, MAE<br>

* MAPE vs. RMSE/ MAE: use MAPE when percentage error is important for your goal<br>

* RMSE vs. MAE: RMSE penalizes very large errors, may favor a model that makes many small errors over a model that makes occasional large errors.

# 4. Regression Trees

* Used with continous outcome variables
* Similar to classification trees (recursive partitioning, pruning)
* Many splits, choose the one that **minimizes impurity**

* Same advantages/limitations like classification trees<br>
** Easy to intrepret, built-in variable selection/reduction<br>
** May not capture more complex interactions between several variables or structure in the data<br>
<br>
* Difference from classification trees:<br>
** Prediction is **average** of numerical target var (classification tree it is majority vote)<br>
** Impurity measured by **sum of squared deviations** from leaf mean<br>
** Performance measured by RMSE
